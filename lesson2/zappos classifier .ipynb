We have to make 4 folders with images inside them without subfolders as it is now 

```python
import os
from pathlib import Path
import shutil

def collapse_files(parent_folder, path = os.getcwd()):
    cur_path = os.getcwd()
    os.chdir(path)
    #Create list of paths to images
    look_path = [parent_folder]
    res = []
    h = look_path
    while len(h) > 0:
        h = []
        for p in look_path:
            #check on last subfolder in tree -> then add to list 
            if( os.path.isfile(p + '/' + os.listdir(p)[0]) or len(os.listdir(p)) == 0 ):
                res.append(p)
            #else add all subfolders in list_path
            else:
                h = h + [p + '/' + x for x in os.listdir(p)]
        look_path = h
    #Copy to the parent folder 
    sub_ = os.listdir(parent_folder)
    for p in res:
        files = os.listdir(p)
        id_ = 1
        for f in files:
            #shutil.copyfile(p + '/' + f, parent_folder + '/' + p[len(p) - p[::-1].find('/'):] + '_' + f)
            shutil.copyfile(p + '/' + f, parent_folder + '/' + str(id_) + '.' + f.split(".")[-1] )
            id_ = id_ + 1 
    for x in sub_:
        shutil.rmtree(parent_folder + '/' + x)
    os.chdir(cur_path)
```

```python
os.listdir('ut-zap50k-images-square')
```

```python
for x in os.listdir(os.getcwd() + '/' + 'ut-zap50k-images-square'):
    collapse_files(x, path = os.getcwd() + '/' + 'ut-zap50k-images-square')
```

### View Data

```python
path = Path('ut-zap50k-images-square')
```

```python
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.2,
        ds_tfms=get_transforms(), size=224, num_workers=4).normalize(imagenet_stats)
```

```python
data.classes
```

```python
data.show_batch(rows=3, figsize=(7,8))
```

```python
data.classes, data.c, len(data.train_ds), len(data.valid_ds)
```

## Train

```python
learn = create_cnn(data, models.resnet34, metrics=error_rate)
```

```python
learn.fit_one_cycle(4)
```

```python
learn.save('stage-1')
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_confusion_matrix()
```
